```python
import matplotlib.pyplot as plt

# Data penjualan (dalam juta rupiah)
bulan = ["Jan", "Feb", "Mar", "Apr", "Mei", "Jun"]
penjualan = [30, 80, 85, 90, 95, 85]

# Untuk membuat grafik garis
plt.figure(figsize=(8,5))
plt.plot(bulan, penjualan, marker='o', color='blue', label="Penjualan")

# Untuk menambahkan judul dan label
plt.title("Data Penjualan Bulanan")
plt.xlabel("Bulan")
plt.ylabel("Penjualan (juta rupiah)")
plt.legend()
plt.grid(True)

# Untuk menampilkan grafik
plt.show()

# Untuk membuat grafik batang
plt.figure(figsize=(8,5))
plt.bar(bulan, penjualan, color='orange')

plt.title("Data Penjualan Bulanan (Bar Chart)")
plt.xlabel("Bulan")
plt.ylabel("Penjualan (juta rupiah)")
plt.show()
```
